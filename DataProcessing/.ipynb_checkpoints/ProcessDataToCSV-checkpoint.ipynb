```python
import numpy as np
import pandas as pd
```

```python
arff_filename = "../Data/chronic_kidney_disease_full.arff"
arff_file = open(arff_filename, "r")
arff_lines = arff_file.readlines()
colnames = []
data_lines = []
df = pd.DataFrame()
count = 0


for line in arff_lines:
    
    # Read in column names from arff file
    if (line[0] == "@"):
        # This takes only the 25 column names and ignores anything else
        if(count > 0 and count <= 25):
            colnames.append((line.split()[1])[1:-1])
        count += 1
        
    # This sets the df column names once we collect them all
    elif (count == 26):
        df = pd.DataFrame(columns= colnames)
        count += 1
    
    # This part reads in the actual data section
    elif(count > 26):
        
        # If the line has data, process and add to df (means it does not start with a space)
        if (not ((line[0]).isspace())):
            
            # Split the line by commas
            data_line = line.split(',')
            # Remove the newline character at the end
            data_line[-1] = data_line[-1].split()
            
            # Accidentally put last entry in its own list, so this takes it out 
            for x in data_line[-1]:
                data_line[-1] = x
             
            # Sometimes the arff line had a space at the end, so this removes that as an entry
            if not data_line[-1]:
                data_line = data_line[0:25]
            
            # Removes any entries that are spaces throughout
            data_line = [i for i in data_line if i]

            # Add the data to the dataframe if the data is the right size
            if (len(data_line) == 25):
                df.loc[count - 27] = data_line
                count += 1
            
            # This should not get printed, but prints the data if it is not correct format
            else:
                print(count - 27)
                print(data_line)
                print(len(data_line))
```

```python
# Process dataframe so its entirely numeric

# Replace ? with None in dataframe

# Process dataframe so:

# Normal = 0
# Abnormal = 1
df = df.replace({'normal': 0})
df = df.replace({'abnormal': 1})

# Not present = 0
# Present = 1
df = df.replace({'notpresent': 0})
df = df.replace({'present': 1})

# No = 0
# Yes = 1
df = df.replace({'no': 0})
df = df.replace({'yes': 1})


# Good = 0
# Poor = 1
df = df.replace({'good': 0})
df = df.replace({'poor': 1})

# Not CKD = 0
# CKD = 1
df = df.replace({'notckd': 0})
df = df.replace({'ckd': 1})

df = df.replace({'?': float('nan')})
df = df.replace({'': float('nan')})
df = df.replace({None: float('nan')})
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
# Write CSV
csv_filename = "../Data/chronic_kidney_disease.csv"
df.to_csv(csv_filename, index = False)
```
